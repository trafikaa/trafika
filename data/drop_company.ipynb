```python
import pandas as pd
import numpy as np
import os
```

```python
# 1. CSV 파일 불러오기
df = pd.read_csv('delisted_financials_all.csv')
```

```python
# 2. statement_name이 '현금흐름표'인 행 제외
df = df[df['statement_name'] != '현금흐름표']

# 3. account_num 칼럼 삭제
if 'account_num' in df.columns:
    df = df.drop(columns=['account_num'])

# 4. ticker로 그룹화, account_id를 column으로 pivot
pivot_df = df.pivot_table(
    index='ticker',
    columns='account_id',
    values='this_term_amount',  # 실제 값 컬럼명으로 변경
    aggfunc='first'  # 여러 값이 있으면 첫 번째 값 사용
)

# 결과 확인
print(pivot_df.head())
```

```python
pivot_df = pivot_df.drop(columns=['ifrs_Cash', 'ifrs_CashOnHand', 'ifrs_BalancesWithBanks', 'ifrs_CashEquivalents', 'ifrs_ShorttermInvestmentsClassifiedAsCashEquivalents'])
```

```python
# 분자: ifrs-full_Assets 또는 ifrs_Assets
assets = None
if 'ifrs-full_Assets' in pivot_df.columns:
    assets = pivot_df['ifrs-full_Assets']
elif 'ifrs_Assets' in pivot_df.columns:
    assets = pivot_df['ifrs_Assets']
else:
    print('자산 계정이 없습니다.')

# 분모: ifrs-full_CurrentLiabilities 또는 ifrs_CurrentLiabilities
current_liabilities = None
if 'ifrs-full_CurrentLiabilities' in pivot_df.columns:
    current_liabilities = pivot_df['ifrs-full_CurrentLiabilities']
elif 'ifrs_CurrentLiabilities' in pivot_df.columns:
    current_liabilities = pivot_df['ifrs_CurrentLiabilities']
else:
    print('유동부채 계정이 없습니다.')

# current_ratio 계산 및 컬럼 추가
if (assets is not None) and (current_liabilities is not None):
    pivot_df['current_ratio'] = assets / current_liabilities
else:
    pivot_df['current_ratio'] = np.nan  # 둘 다 없으면 NaN

# 결과 확인
print(pivot_df[['current_ratio']].head())
```
