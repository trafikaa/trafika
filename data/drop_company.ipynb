```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('delisted_financials_all.csv')

# 필요한 계정만 추출
accounts = [
    'ifrs-full_CurrentAssets', 'ifrs_CurrentAssets',
    'ifrs-full_CurrentLiabilities', 'ifrs_CurrentLiabilities',
    'ifrs-full_Assets', 'ifrs_Assets',
    'ifrs-full_Liabilities', 'ifrs_Liabilities',
    'ifrs-full_ProfitLoss', 'ifrs_ProfitLoss',
    'ifrs-full_Equity', 'ifrs_Equity',
    'ifrs-full_Revenue', 'ifrs_Revenue'
]

df_filtered = df[df['account_id'].isin(accounts)]

# 피벗: ticker, year별로 계정별 this_term_amount를 컬럼으로
pivot = df_filtered.pivot_table(
    index=['ticker', 'year'],
    columns='account_id',
    values='this_term_amount',
    aggfunc='first'
).reset_index()
def coalesce(*cols):
    for col in cols:
        if isinstance(col, pd.Series):
            col = col.iloc[0]
        if col is not None and pd.notnull(col):
            return col
    return None

pivot['CurrentAssets'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_CurrentAssets'), x.get('ifrs_CurrentAssets')), axis=1)
pivot['CurrentLiabilities'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_CurrentLiabilities'), x.get('ifrs_CurrentLiabilities')), axis=1)
pivot['Assets'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_Assets'), x.get('ifrs_Assets')), axis=1)
pivot['Liabilities'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_Liabilities'), x.get('ifrs_Liabilities')), axis=1)
pivot['ProfitLoss'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_ProfitLoss'), x.get('ifrs_ProfitLoss')), axis=1)
pivot['Equity'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_Equity'), x.get('ifrs_Equity')), axis=1)
pivot['Revenue'] = pivot.apply(lambda x: coalesce(x.get('ifrs-full_Revenue'), x.get('ifrs_Revenue')), axis=1)

# 1. 유동비율
pivot['current_ratio'] = pivot['CurrentAssets'] / pivot['CurrentLiabilities']

# 2. 부채비율
pivot['debt_ratio'] = pivot['Assets'] / pivot['Liabilities']

# 3. ROA
pivot['ROA'] = pivot['ProfitLoss'] / pivot['Assets']

# 4. ROE
pivot['ROE'] = pivot['ProfitLoss'] / pivot['Equity']

# 5. 총자산회전율
pivot['asset_turnover'] = pivot['Revenue'] / pivot['Assets']

# 6. 매출증가율 (각 행의 this_term_amount, prev_term_amount로 계산)
revenue_df = df_filtered[df_filtered['account_id'].isin(['ifrs-full_Revenue', 'ifrs_Revenue'])].copy()
revenue_df['revenue_growth'] = (revenue_df['this_term_amount'] - revenue_df['prev_term_amount']) / revenue_df['prev_term_amount']

# 7. 총자산증가율 (각 행의 this_term_amount, prev_term_amount로 계산)
asset_df = df_filtered[df_filtered['account_id'].isin(['ifrs-full_Assets', 'ifrs_Assets'])].copy()
asset_df['asset_growth'] = (asset_df['this_term_amount'] - asset_df['prev_term_amount']) / asset_df['prev_term_amount']

# ticker, year별로 증가율만 추출
revenue_growth = revenue_df[['ticker', 'year', 'revenue_growth']]
asset_growth = asset_df[['ticker', 'year', 'asset_growth']]

# pivot에 merge
result = pd.merge(pivot, revenue_growth, on=['ticker', 'year'], how='left')
result = pd.merge(result, asset_growth, on=['ticker', 'year'], how='left')

# 계산된 지표만 추출
result['ticker'] = result['ticker'].astype(str).str.zfill(6)
result = result[['ticker', 'year', 'current_ratio', 'debt_ratio', 'ROA', 'ROE', 'asset_turnover', 'revenue_growth', 'asset_growth']]

# 결과 저장
result.to_csv('delisted_financials_with_metrics.csv', index=False)
```
